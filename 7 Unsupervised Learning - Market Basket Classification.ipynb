## Importing Requires Libraries

```python
import numpy as np # Array manipulation and linear algebra
import pandas as pd # Creating dataframes and manipulating data
from matplotlib import pyplot as plt # Visualising Data
import seaborn as sns # Visulaising data

from sklearn.preprocessing import MinMaxScaler # Normalising data to show from 0-1
from sklearn.cluster import KMeans # Clustering using Mean
from sklearn_extra.cluster import KMedoids # Clustering using Median

from sklearn.metrics import silhouette_score # Metrics to check how good a cluster is
```

## Importing Dataset

```python
df = pd.read_csv('./Data/Mall_customers.csv')
df.head()
```

## Exploratory Data Analysis

```python
df.isnull().sum()
```

Luckily, there are no null values that must be imputed.

### Managing useful and useless columns

We encode 'Gender' using one-hot encoding. It helps the clustering algorithms work on the data. Here, 1 represents male and 0 represents female. We also drop the 'CustomerID' column because it does not help our cause.

```python
df['Gender'] = pd.get_dummies(df['Gender'], drop_first=True)
df.drop(columns=['CustomerID'], inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='Greens_r')
```

It turns out that the only correlation we can see is between 'Age' and 'Spending Score'. It suggests that with age, the spending score seems to drop. Other than that, the data point are pretty much not related, or so we see.

### Normalising Data

It is very important to normalise the data before clustering. Doing so, we make sure that all our data points have an equal impact on the clusteirng of our values. This is simply because clustering works on the concept of 'distance'. If any column seems to be the cause of the distance to be large, it will end up impacting the data the most. To avoid such a problem, we must make sure to normalise the data.

```python
mms = MinMaxScaler()
data = mms.fit_transform(df)
df = pd.DataFrame(data, index=df.index, columns=df.columns)
df.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='Greens_r')
```

As we can see, the correlations do not change even after normalisation. Thus, normalisation only affects the weights of clustering and nothing else.

```python
sns.pairplot(df)
```

The pairplot shows us that the data is distributed and that we cannot really see a linear correlation. although, we can also see the possiblity of a good clustering when 'Annual Income' and  'Spending Score' are used. Other than that, there is no way to see the presence of any clusters.

## Clustering

### K Means

```python
clusters = 5
randstate = 0
kmn = KMeans(n_clusters = clusters, random_state=randstate)
kmean = kmn.fit(df)
kmean
```

```python
kmean.labels_
```

### K Medoids

```python
kmd = KMedoids(n_clusters = clusters, random_state=randstate)
kmed = kmd.fit(df)
kmed
```

```python
kmed.labels_
```

```python
df['Mean Labels'] = kmean.labels_
df['Medoid Labels'] = kmed.labels_
df.head()
```

## Evaluating and Analysing Clusters

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='Greens_r')
```

```python
sns.scatterplot(x=df['Spending Score (1-100)'], y=df['Age'], hue=df['Mean Labels'], palette='deep')
```

```python
sns.scatterplot(x=df['Gender'], y=df['Spending Score (1-100)'], hue=df['Medoid Labels'], palette='deep')
```

Even when we plot the points with the best correlations, we can see that the clusters are overlapping. This means that the clustering has beenn performed in multiple dimensions and cannot be seen directly with a 2D plot.

```python
silhouette_score(data, kmean.labels_)
```

```python
silhouette_score(data, kmed.labels_)
```

For 5 clusters, K medoids clustering works better than K means.

## Extensive analysis of data clustering

We can extensively analyse the data by plotting the silhoutte score with the number of clusters and identifying the highest score. We can say it is a derivative of the 'elbow' method.

```python
dfn = df
means = []
medos = []

bestmean = []
bestmedo = []

bestscoremean = 0
bestscoremedo = 0

bestmeancluster = 0
bestmedocluster = 0

clusterArr = range(3, 11)
for clusters in clusterArr:
    kmn = KMeans(n_clusters = clusters, random_state=randstate)
    kmean = kmn.fit(dfn)

    kmd = KMedoids(n_clusters = clusters, random_state=randstate)
    kmedo = kmd.fit(dfn)

    meanscore = silhouette_score(dfn, kmean.labels_)
    medoscore = silhouette_score(dfn, kmedo.labels_)
    
    means.append(meanscore)
    medos.append(medoscore)
    
    if meanscore > bestscoremean:
        bestscoremean = meanscore
        bestmean = kmean.labels_
        bestmeancluster = clusters
    
    if medoscore > bestscoremedo:
        bestscoremedo = medoscore
        bestmedo = kmedo.labels_
        bestmedocluster = clusters

dfn['Mean Labels'] = bestmean
dfn['Medoid Labels'] = bestmedo

print(f'Best silhouette score for Kmeans is {bestscoremean} at {bestmeancluster} clusters.')
print(f'Best silhouette score for Kmedoids is {bestscoremedo} at {bestmedocluster} clusters.')
```

As is visible, we get better clusters in K means and the best cluster of K means is better than that of K medoids.

### Plotting K Means Silhoutte Scores

```python
plt.plot(clusterArr, means)
plt.scatter(clusterArr, means)
```

### Plotting K Medoids Silhoutte Scores

```python
plt.plot(clusterArr, medos)
plt.scatter(clusterArr, medos)
```

Even the line plots suggest that the maximum score is the 'elbow' for this  dataset. Thus, we can consider it to be the best cluster using that method.

The pairplot of K Means shows us that even when we plot all the points with each other, we cannot distinguish all the clusters effectively. Thus, the clustering is multidimensional.
