# Applying Clustering on Country data

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
from kmodes.kmodes import KModes

from sklearn.metrics import silhouette_score
```

## Creating a KNN System for K Means Clustering

### Algorithm:

1. Input 'n' as number of centroids and 'k' as the number of nearest neighbours to check from.
2. Select 'n' centroids randomly.
3. Calculate distance of each element from all centroids and assign to closest centroid.
4. Get KNN of the centroid from the new-formed clusters.
5. Generate new centroid by finding the mean of the KNN.
6. If new centroids are same as old centroids, go to step 8.
7. Else, go to step 3.

```python
import random
from copy import deepcopy

def getDistance(centroid, element):
    
    distance = 0
    for i in range(len(centroid)):
        # Manhattan Distance
        distance += abs(centroid[i]-element[i])
    return distance

def getClosest(centroids, element):
    
    closest = 0
    shortest = getDistance(centroids[closest], element)
    for centroid in range(len(centroids)):
        distance = getDistance(centroids[centroid], element)
        if distance < shortest:
            shortest = distance
            closest = centroid
    return closest

def getKNearest(clus, centroid, k):
    
    knn = []
    cluster = deepcopy(clus)
    
    while len(knn) < k and len(cluster) > 0:
        
        closest = 0
        shortest = getDistance(centroid, cluster[closest])
        
        for i in range(len(cluster)):
            dist = getDistance(centroid, cluster[i])
            if dist <= shortest and i != closest:
                closest = i
                shortest = dist
                
        knn.append(cluster[closest])
        del cluster[closest]
    
    return knn

def KNNClustering(n, k, dataframe):
    
    data = []
    datavals = np.array(dataframe)
    for i in range(len(datavals)):
        data.append([])
        for j in range(len(datavals[i])):
            data[i].append(datavals[i][j])

    # Select n centroids randomly
    curCentroids = []
    newCentroids = []
    clusters = []
    clustervalue = []
    datasize = len(data)
    centvals = []
    
    while len(curCentroids) < n:
        cent = random.randint(0, datasize-1)
        if cent not in centvals:
            curCentroids.append(data[cent])
            clusters.append([])
            newCentroids.append([])
            centvals.append(cent)
            
    while newCentroids != curCentroids:
        
        if newCentroids != curCentroids and len(newCentroids[0]) > 0:
            curCentroids = deepcopy(newCentroids)
            
        for i in range(len(clusters)):
            clusters[i] = []
            newCentroids[i] = []
        
        # Calculate distance of each element and assign to the closest centroid
        for i in range(datasize):
            closest = getClosest(curCentroids, data[i])
            clusters[closest].append(data[i])
            if len(clustervalue) == i:
                clustervalue.append(closest)
            else:
                clustervalue[i] = closest
            
        knnclusters = []
        for i in range(len(clusters)):
            knnclusters.append(getKNearest(clusters[i], curCentroids[i], k))
        
        # Calculate the new centriods
        for cluster in range(len(knnclusters)):
            for element in range(len(knnclusters[cluster])):
                for attribute in range(len(knnclusters[cluster][element])):
                    if len(newCentroids[cluster]) == attribute:
                        newCentroids[cluster].append(0)
                    
                    # Calculating mean
                    newCentroids[cluster][attribute] *= element
                    newCentroids[cluster][attribute] += knnclusters[cluster][element][attribute]
                    newCentroids[cluster][attribute] /= element+1
                    
    return [clustervalue, newCentroids]
```

## Importing data

```python
df = pd.read_csv('./Data/Country-data.csv')
df
```

```python
df.isnull().sum()
```

```python
mms = MinMaxScaler()
data = mms.fit_transform(df.drop(columns=['country']))
data
```

```python
country = df['country']
df.drop(columns=['country'], inplace=True)
```

```python
df = pd.DataFrame(data, index=df.index, columns=df.columns)
df['country'] = country
df
```

## Applying Clustering

```python
clusters = 3
randstate = 18
kmn = KMeans(n_clusters = clusters, random_state=randstate)
kmean = kmn.fit(df.drop(columns=['country']))
kmean
```

```python
kmean.labels_
```

```python
kmd = KMedoids(n_clusters = clusters, random_state=randstate)
kmed = kmd.fit(df.drop(columns=['country']))
kmed
```

```python
kmed.labels_
```

```python
df['Mean Labels'] = kmean.labels_
df['Medoid Labels'] = kmed.labels_
df
```

Having taken number of clusters as 3, we see that all 3 clustering algortihms have about the same silhoutte score. Athough, it appears that K Means and K Modes are better when compared to K Medoids and KNN in K Means. Among the 2, K Means has a slighlty higher silhoutte score. This means that the best way to cluster the data into 3 parts would be K Means, followed closely by K Modes which is then followed by K Medoids ending with KNN.

```python
dfn = df
means = []
medos = []
modes = []
knens = []
clusterArr = range(3, 11)

for clusters in clusterArr:
    kmn = KMeans(n_clusters = clusters, random_state=randstate)
    kmean = kmn.fit(dfn.drop(columns=['country']))

    kmd = KMedoids(n_clusters = clusters, random_state=randstate)
    kmedo = kmd.fit(dfn.drop(columns=['country']))

    kmo = KModes(n_clusters = clusters, random_state=randstate)
    kmode = kmo.fit(dfn[['Mean Labels', 'Medoid Labels']])
    
    KNN = KNNClustering(clusters, 10, dfn.drop(columns=['country']))
    
    means.append(silhouette_score(dfn.drop(columns=['country']), kmean.labels_))
    medos.append(silhouette_score(dfn.drop(columns=['country']), kmedo.labels_))
    modes.append(silhouette_score(dfn[['Mean Labels', 'Medoid Labels']], kmean.labels_))
    knens.append(silhouette_score(dfn.drop(columns=['country']), KNN[0]))
```

It appears that the clusters usually get better beyond 3. Also, it seems that the optima keeps shifting wiith every new run. Although, we can say that we find our "elbow" to be almost the same number of clusters for all algorithms. Although it is actually just an optima. The values are fluctuating and thus we do not get any form of a linear shape.

```python
clusters = 4
kmn = KMeans(n_clusters = clusters, random_state=randstate)
kmean = kmn.fit(dfn.drop(columns=['country']))
dfn['Mean Labels'] = kmean.labels_

kmd = KMedoids(n_clusters = clusters, random_state=randstate)
kmedo = kmd.fit(dfn.drop(columns=['country']))
dfn['Medoid Labels'] = kmedo.labels_

kmo = KModes(n_clusters = clusters, random_state=randstate)
kmode = kmo.fit(dfn[['Mean Labels', 'Medoid Labels']])
dfn['Mode Labels'] = kmode.labels_

KNN = KNNClustering(clusters, 10, dfn.drop(columns=['country']))
knncentroids = KNN[1]
dfn['KNN Labels'] = KNN[0]
```
